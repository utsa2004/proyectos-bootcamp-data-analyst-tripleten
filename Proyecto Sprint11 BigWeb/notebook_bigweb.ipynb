### **3.1 GRAFICO DEL INGRESO ACUMULADO POR GRUPO**

</div>

```python
# Preparar estructura de fechas y grupos
datesgroups = orders[['date', 'group']].drop_duplicates()
```

```python
# Calcular m√©tricas acumuladas
# Ordenes acumuladas(por fecha y grupo)
ordersAggregated = datesgroups.apply(
    lambda x: orders[
        (orders['date'] <= x['date']) & (orders['group'] == x['group'])
    ].agg({
        'date' : 'max',
        'group' : 'max',
        'transactionid' : 'nunique',
        'visitorid' : 'nunique',
        'revenue' : 'sum'
    }),
    axis=1
).sort_values(by=['date', 'group'])

# Visitantes Acumulados
visitsAggregated = datesgroups.apply(
    lambda x: visits[
        (visits['date'] <= x['date']) & (visits['group'] == x['group'])
    ].agg({
        'date' : 'max',
        'group' : 'max',
        'visits' : 'sum'
    }),
    axis=1
).sort_values(by=['date', 'group'])
```

```python
# Unir en un solo dataframe orders y visits(Aggregated)
cumulativeData = ordersAggregated.merge(visitsAggregated, on=['date', 'group'])
# Renombrar columnas en el nuevo dataframe creado
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
```

### **3.9 CALCULO DE LA SIGNIFICANCIA ESTADISTICA(VALOR P) DE LA DIFERENCIA EN LA TASA DE CONVERSION ENTRE LOS GRUPOS, UTILIZANDO DATOS EN BRUTO**

</div>

```python
# Agrupar por usuario y contar la cantidad de pedidos √∫nicos en cada grupo
ordersbuyersA = orders[orders['group'] == 'A'].groupby('visitorid')['transactionid'].nunique().reset_index().rename(columns={'transactionid' : 'orders'})
ordersbuyersB = orders[orders['group'] == 'B'].groupby('visitorid')['transactionid'].nunique().reset_index().rename(columns={'transactionid' : 'orders'})    
```

```python

# Crear las muestras agregando usuarios con 0 pedidos
sampleA = pd.concat([
    ordersbuyersA['orders'],
    pd.Series(
        0,
        index=np.arange(
            visits[visits['group'] == 'A']['visits'].sum() - len(ordersbuyersA['orders'])
        ),
        name='orders'
    )
], axis=0)
sampleB = pd.concat([
    ordersbuyersB['orders'],
    pd.Series(
        0,
        index=np.arange(
            visits[visits['group'] == 'B']['visits'].sum() - len(ordersbuyersB['orders'])
        ),
        name='orders'
    )
], axis=0)
```

```python
# Calcular y mostrar la significancia estad√≠stica(valor p) de la diferencia en la tasa de conversi√≥n entre los dos grupos

# Establecemos valor de alpha
alpha = 0.05

print('La significancia estad√≠stica(valor p) de la diferencia en la tasa de conversi√≥n entre los grupos es de:', "{0:.4f}".format(st.mannwhitneyu(sampleA, sampleB)[1]))
```

```python
# Extra, calcular la diferencia relativa en la tasa de conversi√≥n entre los grupos

print('La diferencia relativa en la tasa de conversi√≥n entre los grupos es de:', "{0:.4f}".format(sampleB.mean() / sampleA.mean() - 1))
```

El **valor p** obtenido al comparar la conversi√≥n entre los grupos A y B **es 0.0110**, lo que es menor al umbral de significancia estad√≠stica definido (Œ± = 0.05).

Esto indica que existe una diferencia estad√≠sticamente significativa en la conversi√≥n entre los dos grupos, lo cual permite rechazar la hip√≥tesis nula de que ambos grupos tienen la misma tasa de conversi√≥n.

```python
# Pedidos por usuario por grupo
ordersbyusersA = orders[orders['group'] == 'A'].groupby('visitorid')['transactionid'].nunique().reset_index().rename(columns={'transactionid' : 'orders'})
ordersbyusersB = orders[orders['group'] == 'B'].groupby('visitorid')['transactionid'].nunique().reset_index().rename(columns={'transactionid' : 'orders'})
```

```python
# Usuarios con m√°s de 1 pedido
userswithmanyorders = pd.concat([
    ordersbyusersA[ordersbyusersA['orders'] > 1]['visitorid'],
    ordersbyusersB[ordersbyusersB['orders'] > 1]['visitorid'],
], axis=0)

# Usuarios con pedidos => $415
userswithexpensiveorders = orders[orders['revenue'] > 415]['visitorid']
```

```python
# Usuarios an√≥malos: pedidos >=1 o ingresos => $415
abnormalusers = pd.concat([
    userswithmanyorders,
    userswithexpensiveorders
], axis=0).drop_duplicates().sort_values()
```

```python
# Filtrar datos excluyendo usuarios an√≥malos para el grupo A
sampleAFiltered = pd.concat([
    ordersbyusersA[np.logical_not(ordersbyusersA['visitorid'].isin(abnormalusers))]['orders'],
    pd.Series(0, index=np.arange(
            visits[visits['group'] == 'A']['visits'].sum() - len(ordersbyusersA['orders'])
        ), name='orders')
], axis=0)
```

```python
# Filtrar datos excluyendo usuarios an√≥malos para el grupo B
sampleBFiltered = pd.concat([
    ordersbyusersB[np.logical_not(ordersbyusersB['visitorid'].isin(abnormalusers))]['orders'],
    pd.Series(0, index=np.arange(
            visits[visits['group'] == 'B']['visits'].sum() - len(ordersbyusersB['orders'])
        ), name='orders')
], axis=0)
```

```python
# Calcular y mostrar la significancia estad√≠stica(valor p) de la diferencia en la tasa de conversi√≥n entre los dos grupos, con datos filtrados

# Establecemos valor de alpha
alpha = 0.05

print('La significancia estad√≠stica(valor p) de la diferencia en la tasa de conversi√≥n entre los grupos con datos filtrados es de:', "{0:.4f}".format(st.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
```
