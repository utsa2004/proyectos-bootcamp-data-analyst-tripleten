```python
# Cambiar la naturaleza de los datos de la columna 'date' de object a datetime
calls['date'] = pd.to_datetime(calls['date']).dt.tz_localize(None)
# Convertir columna 'internal' a tipo booleano
calls['internal'] = calls['internal'].astype(bool)
calls.dtypes
```

<div class='alert alert-block alert-warning'><b>Oportunidad de mejora</b> <a class='tocSkip'></a><br>La conversi√≥n de la columna <code>internal</code> con <code>.astype(bool)</code> puede introducir sesgos si su tipo original es <i>object</i> con strings y valores ausentes: en pandas, todo string no vac√≠o (incluido 'False') se convierte en True, y <code>NaN</code> suele evaluarse como True, lo que altera la se√±al. Para una conversi√≥n segura, conviene usar un mapeo expl√≠cito o el tipo booleano nativo nullable:<br><br><code># Ejemplos seguros<br>calls['internal'] = calls['internal'].map({'True': True, 'False': False}).astype('boolean')<br># o, si ya son booleanos/0-1:<br>calls['internal'] = calls['internal'].astype('boolean')</code><br><br>Esto asegura que <code>False</code> permanezca como falso y que los ausentes sigan siendo <code>&lt;NA&gt;</code> hasta que decidas tratarlos. Corregir este punto ayuda a que todas las segmentaciones posteriores (internas/externas) reflejen la realidad operativa.</div>

```python
# Verificar si existen valores duplicados expl√≠citos
duplicate1 = calls.duplicated().sum()
print(f'La cantidad de valores duplicados expl√≠citos del dataframe calls es de: {duplicate1}')
```

Los indicadores analizados corresponden a las siguientes **tres hip√≥tesis**:

1. **Hip√≥tesis 1:** Diferencias en el n√∫mero promedio de **llamadas entrantes perdidas** entre operadores (Secci√≥n 5.1).  
2. **Hip√≥tesis 2:** Diferencias en el **tiempo promedio de espera de llamadas entrantes externas** (Secci√≥n 5.2).  
3. **Hip√≥tesis 3:** Diferencias en el **n√∫mero promedio de llamadas salientes externas** (Secci√≥n 5.3).  

üîç En todos los casos, los **dos grupos de comparaci√≥n** se conforman de la siguiente manera:

- **Grupo 1:** Operadores clasificados como ineficientes (identificados en la Secci√≥n 5 mediante percentiles de desempe√±o).  
- **Grupo 2:** Resto de los operadores (con desempe√±o dentro de los rangos normales).

El uso del **estad√≠stico t de Welch** se justifica debido a que los tama√±os y varianzas de los grupos son desiguales, lo cual invalida el supuesto de homocedasticidad requerido por la t de Student cl√°sica. Por ello, la versi√≥n de Welch (configurada con `equal_var=False`) ofrece una comparaci√≥n m√°s robusta y precisa bajo condiciones de heterogeneidad.

El enfoque metodol√≥gico se centra en:

- Verificar la **igualdad de varianzas** entre los grupos mediante la **prueba de Levene**.  
- Evaluar la **diferencia en las medias** de desempe√±o con la **prueba t de Welch**.  
- Determinar si las diferencias detectadas reflejan **variaciones reales en la eficiencia operativa** o son atribuibles al azar.  
- Traducir los resultados en **conclusiones pr√°cticas e inferenciales** que fortalezcan la validez del diagn√≥stico exploratorio previo.  

üí° Esta fase consolida los an√°lisis desarrollados en la **Secci√≥n 5 (Identificaci√≥n de Operadores Ineficaces)**, validando si los comportamientos observados en las m√©tricas clave ‚Äîllamadas entrantes perdidas (5.1), tiempos de espera de llamadas entrantes externas (5.2) y llamadas salientes externas (5.3)‚Äî presentan diferencias estad√≠sticamente significativas entre los operadores ineficientes y el resto del equipo.

<div style="background-color:#EFEBD6; padding:15px; border-radius:12px;">

### **6.1 HIPOTESIS 1: LLAMADAS ENTRANTES PERDIDAS**
- **H‚ÇÄ:** No existen diferencias significativas en el n√∫mero promedio de llamadas entrantes perdidas entre los operadores.  
- **H‚ÇÅ:** Existen diferencias significativas en el n√∫mero promedio de llamadas entrantes perdidas entre los operadores.  

</div>


```python
# Eliminar filas sin operator_id (no pueden asociarse a ning√∫n operador)
telfusion = telfusion.dropna(subset=['operator_id'])

# Convertir a entero para consistencia (por si a√∫n es float)
telfusion['operator_id'] = telfusion['operator_id'].astype(int)

# Informaci√≥n del dataframe 'telfusion'
telfusion.info()
```

```python
# Filtrar llamadas entrantes
in_calls = telfusion[telfusion['direction'] == 'in']

# Calcular el n√∫mero total de llamadas entrantes perdidas por operador
missed_calls = (
    in_calls[in_calls['is_missed_call'] == True]
    .groupby('operator_id')['is_missed_call']
    .count()
    .reset_index()
    .rename(columns={'is_missed_call': 'missed_call_count'})
)
```

```python
# Determinar el umbral del percentil 90 (operadores ineficientes)
umbral_ineficiencia = missed_calls['missed_call_count'].quantile(0.90)
print(f"üîπ Umbral del percentil 90: {umbral_ineficiencia:.2f} llamadas perdidas")

# Clasificar los operadores seg√∫n el umbral
ineficientes = missed_calls[missed_calls['missed_call_count'] > umbral_ineficiencia]
resto = missed_calls[missed_calls['missed_call_count'] <= umbral_ineficiencia]

print(f"Operadores ineficientes: {len(ineficientes)}")
print(f"Resto de operadores: {len(resto)}")
```

```python
# üîπ PRUEBA DE LEVENE (igualdad de varianzas)
stat_levene, p_levene = st.levene(ineficientes['missed_call_count'],
                                 resto['missed_call_count'],
                                 center='mean')

print("üß™ Resultado prueba de Levene (igualdad de varianzas):")
print(f"Estad√≠stico = {stat_levene:.4f}, p-value = {p_levene:.4f}")

if p_levene > 0.05:
    print("‚úÖ No se rechaza H‚ÇÄ ‚Üí Varianzas homog√©neas entre los grupos.")
    equal_var = True
else:
    print("‚ö†Ô∏è Se rechaza H‚ÇÄ ‚Üí Varianzas heterog√©neas entre los grupos.")
    equal_var = False
```

```python
# üîπ PRUEBA T DE WELCH (bilateral, varianzas desiguales)
stat_t, p_t = st.ttest_ind(ineficientes['missed_call_count'],
                           resto['missed_call_count'],
                           equal_var=False)  # <- Welch

print("üß™ Resultado prueba t de Welch (comparaci√≥n de medias):")
print(f"Estad√≠stico t = {stat_t:.4f}, p-value = {p_t:.4f}")

# Nivel de significancia
alpha = 0.05

if p_t < alpha:
    print("\nüìâ Se rechaza H‚ÇÄ ‚Üí Existen diferencias estad√≠sticamente significativas en el n√∫mero promedio de llamadas entrantes perdidas entre operadores ineficientes y el resto.")
else:
    print("\nüìà No se rechaza H‚ÇÄ ‚Üí No se encontraron diferencias estad√≠sticamente significativas en el n√∫mero promedio de llamadas entrantes perdidas entre los grupos.")
```

<div class='alert alert-block alert-success'><b>Acierto o fortaleza</b> <a class='tocSkip'></a><br>Muy bien planteado el flujo inferencial: comprobar varianzas con Levene y luego usar la t de Welch cuando corresponde es un est√°ndar robusto. Para enriquecer la interpretaci√≥n pr√°ctica, suele aportar reportar tambi√©n un tama√±o de efecto (p. ej., d de Cohen) y, si es posible, intervalos de confianza de la diferencia de medias.</div>

Esto implica que el supuesto de homogeneidad de varianzas no se cumple, justificando as√≠ el uso del **estad√≠stico t de Welch**, que corrige este tipo de desigualdad.
