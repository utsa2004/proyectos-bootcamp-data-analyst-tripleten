### **üîπ 4.2. ENCUENTRA EL NUMERO DE RESE√ëAS DE USUARIOS Y LA CALIFICACION PROMEDIO DE CADA LIBRO**    

</div>

```python
query = """
SELECT
    ratings.book_id AS Book_ID,
    COUNT(reviews.review_id) AS Total_Reviews,
    ROUND(AVG(ratings.rating), 2) AS Average_Book_Rating
FROM
    ratings
    LEFT JOIN reviews ON ratings.book_id = reviews.book_id
GROUP BY
    ratings.book_id
ORDER BY
    Average_Book_Rating DESC;
"""
ratings_and_reviews = pd.read_sql(query, con=engine)
display(ratings_and_reviews)
```

<div class="alert alert-block alert-success">
<b>Acierto o fortaleza</b> <a class="tocSkip"></a><br>
Las conclusiones que acompa√±an a [10] sintetizan bien el objetivo y el criterio del <code>LEFT JOIN</code> (incluir libros con calificaci√≥n aunque no tengan rese√±as). Esta claridad metodol√≥gica ayuda a quien lee a entender por qu√© uniones de este tipo son √∫tiles para no perder registros relevantes en an√°lisis descriptivos.
</div>

<div style="background-color: #EAD6D1; padding: 15px; border-radius: 10px;">
    
#### **CONCLUSIONES 4.2: ENCUENTRA EL NUMERO DE RESE√ëAS DE USUARIOS Y LA CALIFICACION PROMEDIO DE CADA LIBRO**

A partir del an√°lisis de las tablas <i>ratings</i> y <i>reviews</i>, se obtuvo el n√∫mero total de rese√±as registradas y la calificaci√≥n promedio correspondiente a cada libro. 
Para asegurar que todos los libros con calificaci√≥n fueran incluidos, se utiliz√≥ un <code>LEFT JOIN</code>, lo que permite incorporar aquellos t√≠tulos que poseen valoraciones pero no necesariamente comentarios asociados. <br><br>
En total se identificaron <b>1,000 libros</b> con calificaciones registradas. El uso de la funci√≥n agregada <code>AVG()</code> permiti√≥ calcular la media de las valoraciones por cada identificador de libro (<i>book_id</i>), y con <code>COUNT()</code> se determin√≥ el n√∫mero total de rese√±as asociadas.

### **üîπ 4.3. IDENTIFICA LA EDITORIAL QUE HA PUBLICADO EL MAYOR NUMERO DE LIBROS CON MAS DE 50 PAGINAS**    

</div>

```python
query = """
SELECT
    books.publisher_id AS Publisher_Id,
    publishers.publisher AS Publisher_Name,
    COUNT(books.book_id) AS Books_Over_50_Pages
FROM
    books
    LEFT JOIN publishers ON books.publisher_id = publishers.publisher_id
WHERE
    books.num_pages > 50
GROUP BY
    books.publisher_id,
    publishers.publisher
ORDER BY
    Books_Over_50_Pages DESC;
"""
top_publishers = pd.read_sql(query, con=engine)
display(top_publishers)
```

### **üîπ 4.4. IDENTIFICA AL AUTOR QUE TIENE LA MAS ALTA CALIFICACION PROMEDIO DEL LIBRO**    

</div>

```python
query = """
SELECT
    authors.author_id AS Author_id,
    authors.author AS Author_Name,
    ROUND(AVG(ratings.rating),2) AS Rating_Book_Average
FROM
    ratings
    LEFT JOIN books ON ratings.book_id = books.book_id
    LEFT JOIN authors ON books.author_id = authors.author_id
GROUP BY
    authors.author_id,
    authors.author
HAVING
    COUNT(ratings.rating) >= 50
ORDER BY
    Rating_Book_Average DESC
LIMIT 10;
"""
top_author_by_rating_book = pd.read_sql(query, con=engine)
top_author_by_rating_book
```

<div class="alert alert-block alert-success">
<b>Acierto o fortaleza</b> <a class="tocSkip"></a><br>
En la celda [12], el uso de <code>HAVING COUNT(ratings.rating) &gt;= 50</code> para asegurar un m√≠nimo de evaluaciones por autor es un gran acierto. Esta decisi√≥n reduce el sesgo por tama√±os de muestra peque√±os y fortalece la validez de la comparaci√≥n entre autores. Adem√°s, el redondeo y el <code>LIMIT 10</code> cierran bien el reporte.

### **üîπ 4.5. NUMERO PROMEDIO DE RESE√ëAS ENTRE LOS USUARIOS QUE CALIFICARON MAS DE 50 LIBROS**    

</div>

```python
# Obtener la lista de usuarios que calificaron m√°s de 50 libros.
query = """
SELECT
    ratings.username AS Username,
    COUNT(reviews.review_id) AS Total_Reviews
FROM
    ratings
LEFT JOIN reviews ON ratings.username = reviews.username
GROUP BY
    ratings.username
HAVING
    COUNT(ratings.rating) > 50
ORDER BY
    Total_Reviews DESC;
"""

top_review_users = pd.read_sql(query, con=engine)
display(top_review_users)
```

```python
## Realizar la subconsulta: Promedio de rese√±as entre los usuarios que calificaron m√°s de 50 libros
query_avg = """
SELECT
    ROUND(AVG(user_reviews.Total_Reviews), 2) AS Avg_Reviews_Among_Top_Users
FROM (
    SELECT
        ratings.username,
        COUNT(reviews.review_id) AS Total_Reviews
    FROM
        ratings
    LEFT JOIN reviews ON ratings.username = reviews.username
    GROUP BY
        ratings.username
    HAVING
        COUNT(ratings.rating) > 50
) AS user_reviews;
"""

avg_reviews = pd.read_sql(query_avg, con=engine)
display(avg_reviews)

```
