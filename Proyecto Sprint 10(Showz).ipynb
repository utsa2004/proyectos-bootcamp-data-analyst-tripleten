#### **2.1A. CUANTAS PERSONAS VISITAN LA PAGINA: DIA, SEMANA Y MES**

</div>

```python
# Convertir columna 'start_ts' a datetime
visits['start_ts'] = pd.to_datetime(visits['start_ts'])

# Crear las columnas esperadas por la plataforma
visits['session_year'] = visits['start_ts'].dt.isocalendar().year
visits['session_month'] = visits['start_ts'].dt.month
visits['session_week'] = visits['start_ts'].dt.isocalendar().week
visits['session_date'] = visits['start_ts'].dt.date

# C√°lculo de los usuarios activos diarios(dau) para todo el per√≠odo
dau_total = (
    visits.groupby('session_date')
    .agg({'uid': 'nunique'})
    .mean()
)

# C√°lculo de los usuarios activos semanales(wau) para todo el per√≠odo
wau_total = (
    visits.groupby(['session_year', 'session_week'])
    .agg({'uid': 'nunique'})
    .mean()
)
# C√°lculo de los usuarios activos mensuales(mau) para todo el per√≠odo
mau_total = (
    visits.groupby(['session_year', 'session_month'])
    .agg({'uid': 'nunique'})
    .mean()
)

print(f"El total de usuarios activos diarios(DAU) es de: {dau_total[0]:,.2f}")
print(f"El total de usuarios activos semanales(WAU) es de: {wau_total[0]:,.2f}")
print(f"El total de usuarios activos mensuales(MAU) es de: {mau_total[0]:,.2f}")
```

```python
# N√∫mero de sesiones y de usuarios por d√≠a
sessions_per_user = visits.groupby(['session_date']).agg({'uid' : ['count', 'nunique']})
# Renombrar columna
sessions_per_user.columns = ['n_sessions', 'n_users']
# C√°lculo del n√∫mero de sesiones diarias por usuario
sessions_per_user['sess_per_user'] = sessions_per_user['n_sessions'] / sessions_per_user['n_users']
# Calcular promedio general de sesiones por usuario por d√≠a
avg_sess_per_user = sessions_per_user['sess_per_user'].mean()
print(f"Promedio de sesiones por usuario por d√≠a: {avg_sess_per_user:.2f}")

# N√∫mero de sesiones y de usuarios por mes (METRICA ADICIONAL)
sessions_per_user_m = visits.groupby(['session_year', 'session_month']).agg({'uid' : ['count', 'nunique']})
# Renombrar columna
sessions_per_user_m.columns = ['n_sessions', 'n_users']
# C√°lculo del n√∫mero de sesiones mensuales por usuario
sessions_per_user_m['sess_per_user'] = sessions_per_user_m['n_sessions'] / sessions_per_user_m['n_users']
# Calcular promedio general de sesiones por usuario por mes
avg_sess_per_user_m = sessions_per_user_m['sess_per_user'].mean()
print(f"Promedio de sesiones por usuario por mes: {avg_sess_per_user_m:.2f}")
```

#### **2.2B. CUANTOS PEDIDOS SE REALIZAN AL MES**

```python

# Agrupar por cliente y obtener su primera fecha de compra
first_order_date = orders.groupby('uid')['buy_ts'].min()

# Renombrar la columna de la serie
first_order_date.name = 'first_order_date'

# Unir la serie con el dataframe original
purchases = orders.join(first_order_date, on='uid')

# Convertir las fechas a tipo datetime y extraer el a√±o-mes.
purchases['purchase_month'] = purchases['buy_ts'].astype('datetime64[M]')
purchases['first_purchase_month'] = purchases['first_order_date'].astype('datetime64[M]')

# Encuentra el n√∫mero de compras y el n√∫mero de usuarios que realizaron compras
purchases_grouped = purchases.groupby(['first_purchase_month', 'purchase_month']).agg({
    'buy_ts' : 'nunique',
    'uid' : 'nunique'
})
# Crea una columna 'purchases_per_user', que calcula el promedio de compras por usuario
purchases_grouped['purchases_per_user'] = purchases_grouped['buy_ts'] / purchases_grouped['uid'] 

# Compila una tabla din√°mica
mean_purchases_pivot  = purchases_grouped.pivot_table(
    index = 'first_purchase_month',
    columns = 'purchase_month',
    values = 'purchases_per_user',
    aggfunc = 'mean'
)

# Formatear √≠ndices y columnas de la tabla para que se vean como 'YYYY-MM'
mean_purchases_pivot.index = mean_purchases_pivot.index.strftime('%Y-%m')
mean_purchases_pivot.columns = mean_purchases_pivot.columns.strftime('%Y-%m')

# Muestra la tabla din√°mica (Opcional)
# print(mean_purchases_pivot) # Ya no es necesario imprimirla si usamos heatmap

# Establecer el tama√±o del gr√°fico
plt.figure(figsize=(8, 4))

# Creaci√≥n de Gr√°fico de Mapa de Color
plt.figure(figsize=(8,4.5))
sns.heatmap(mean_purchases_pivot, annot=True, fmt='.2f', cmap='Blues')
plt.title("Promedio de Compras Por Usuario Seg√∫n Cohorte y Mes de Compra", fontsize=14)
plt.ylabel("Mes de Primera Compra", fontsize=12)
plt.xlabel("Mes de Compra", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Agrupar por cliente y obtener su primera fecha de compra
first_order_date = orders.groupby('uid')['buy_ts'].min()
first_order_date.name = 'first_order_date'

# Unir la serie con el dataframe original
purchases = orders.join(first_order_date, on='uid')

# Crear columnas de a√±o-mes para cohortes y mes de compra
purchases['purchase_month'] = purchases['buy_ts'].astype('datetime64[M]')
purchases['first_purchase_month'] = purchases['first_order_date'].astype('datetime64[M]')

# Agrupar por cohorte y mes de compra
purchases_grouped = purchases.groupby(['first_purchase_month', 'purchase_month']).agg({
    'revenue': 'sum',
    'buy_ts': 'count',
    'uid': 'nunique'  
}).reset_index()

# Calcular m√©tricas
purchases_grouped['avg_order_size'] = purchases_grouped['revenue'] / purchases_grouped['buy_ts']
purchases_grouped['revenue_per_user'] = purchases_grouped['revenue'] / purchases_grouped['uid']

# Crear tabla din√°mica - Tama√±o promedio de compra
avg_order_size_pivot = purchases_grouped.pivot_table(
    index='first_purchase_month',
    columns='purchase_month',
    values='avg_order_size',
    aggfunc='mean'
)

# Crear tabla din√°mica - Revenue por usuario
revenue_per_user_pivot = purchases_grouped.pivot_table(
    index='first_purchase_month',
    columns='purchase_month',
    values='revenue_per_user',
    aggfunc='mean'
)

# Formatear √≠ndices para visualizaci√≥n
avg_order_size_pivot.index = avg_order_size_pivot.index.strftime('%Y-%m')
avg_order_size_pivot.columns = avg_order_size_pivot.columns.strftime('%Y-%m')
revenue_per_user_pivot.index = revenue_per_user_pivot.index.strftime('%Y-%m')
revenue_per_user_pivot.columns = revenue_per_user_pivot.columns.strftime('%Y-%m')

# --- Gr√°ficos ---

# Mapa de Calor 1 - Tama√±o promedio de compra
plt.figure(figsize=(8, 4))
sns.heatmap(avg_order_size_pivot, annot=True, fmt='.2f', cmap='Greens')
plt.title("Tama√±o Promedio de Compra por Cohorte y Mes", fontsize=14)
plt.ylabel("Mes de Primera Compra", fontsize=12)
plt.xlabel("Mes de Compra", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Mapa de Calor 2 - Ingresos promedio por usuario
plt.figure(figsize=(8, 4.))
sns.heatmap(revenue_per_user_pivot, annot=True, fmt='.2f', cmap='Oranges')
plt.title("Ingresos Promedio por Usuario por Cohorte y Mes", fontsize=14)
plt.ylabel("Mes de Primera Compra", fontsize=12)
plt.xlabel("Mes de Compra", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

#### **2.2D. DINERO QUE GENERAN LOS CLIENTES (LTV)**

```python
# C√°lculo del margin rate 
margin_rate = round(((orders['revenue'].sum() - costs['costs'].sum()) / orders['revenue'].sum()) * 100, 2)

print(f"El margen de utilidad **bruta** (ingresos - costos variables) de la plataforma Showz es de: {margin_rate}%")
```

<div class="alert alert-block alert-warning">
<b>Celda [34]</b> <a class="tocSkip"></a><br>
El c√°lculo del margen de utilidad bruta es un buen inicio para evaluar la rentabilidad de la plataforma. Considera desglosar los costos para identificar qu√© componentes est√°n afectando m√°s el margen. Esto te permitir√° identificar √°reas espec√≠ficas de mejora y hacer ajustes que podr√≠an mejorar la rentabilidad general.
</div>


```python
# Agrupar por cliente y obtener su primera fecha de compra
first_orders = orders.groupby('uid').agg({'buy_ts': 'min'})

# Renombrar la serie
first_orders.columns = ['first_order_month']

# Agrupar las fechas a tipo datetime y extraer a√±o-mes
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
first_orders['first_order_month'] = first_orders['first_order_month'].astype('datetime64[M]')

#Unir la serie al dataframe original
orders_merged = orders.join(first_orders, on='uid')

#C√°lculo de nuevos clientes para cada mes
cohort_sizes = (
    orders_merged.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']

# Calcular el revenue total generado por cohorte
cohorts = orders_merged.groupby(['first_order_month', 'order_month']).agg({'revenue' : 'sum'}).reset_index()

cohorts.columns = ['first_order_month', 'order_month', 'cohort_revenue']

# fusiona las cohortes y cohort_sizes
report = pd.merge(cohorts, cohort_sizes, on='first_order_month')

# encuentra la edad de cada cohorte en meses
report['age'] = report['order_month'] - report['first_order_month']
report['age'] = report['age'] / np.timedelta64(1,'M')
report['age'] = report['age'].round().astype(int) 

# calcula el LTV para las cohortes
report['ltv'] = report['cohort_revenue'] / report['n_buyers']

# calcula los valores para la tabla din√°mica
result = report.pivot_table(
    index='first_order_month',
    columns='age',
    values='ltv',
    aggfunc='mean'
).round()
result = result.fillna('')

# Imprimir la tabla din√°mica
print(result)
```

```python
# Unir 'orders' con 'visits' por 'uid' para conocer la fuente de cada cliente
orders_with_source = orders.merge(visits[['uid', 'source_id']], on='uid', how='left')

# Contar usuarios √∫nicos que compraron por fuente
buyers_per_source = orders_with_source.groupby('source_id')['uid'].nunique().reset_index()
buyers_per_source.columns = ['source_id', 'buyers']

# Sumar el costo total por fuente
total_costs_per_source = costs.groupby('source_id')['costs'].sum().reset_index()
total_costs_per_source.columns = ['source_id', 'total_cost']

# Unir ambos para calcular el CAC
cac = total_costs_per_source.merge(buyers_per_source, on='source_id')
cac['CAC'] = cac['total_cost'] / cac['buyers']

# Mostrar resultados ordenados
cac = cac.sort_values(by='CAC', ascending=False)
print(cac)
```

#### **2.3C. RENTABILIDAD DE LA INVERSION POR FUENTE (ROMI)**

</div>

```python
# C√°lculo de ROMI por fuente de adquisici√≥n con enfoque de econom√≠a unitaria

# Unir orders con visits para identificar la fuente de cada compra
merged = orders.merge(visits, on='uid')

# Calcular ingresos (revenue) totales por fuente
revenue_by_source = merged.groupby('source_id')['revenue'].sum().reset_index()

# Calcular n√∫mero de compradores √∫nicos por fuente
buyers_by_source = merged.groupby('source_id')['uid'].nunique().reset_index()
buyers_by_source.columns = ['source_id', 'buyers']

# Calcular LTV por fuente (sin usar margen)
ltv = revenue_by_source.merge(buyers_by_source, on='source_id')
ltv['LTV'] = ltv['revenue'] / ltv['buyers']

# Calcular CAC por fuente (costos / compradores)
costs_by_source = costs.groupby('source_id')['costs'].sum().reset_index()
costs_by_source.columns = ['source_id', 'total_cost']

cac = costs_by_source.merge(buyers_by_source, on='source_id')
cac['CAC'] = cac['total_cost'] / cac['buyers']

# Unir LTV y CAC para calcular ROMI
romi = ltv.merge(cac[['source_id', 'CAC']], on='source_id')
romi['ROMI'] = romi['LTV'] / romi['CAC']

# Ordenar resultados por ROMI descendente
romi = romi.sort_values(by='ROMI', ascending=False).reset_index(drop=True)

# Mostrar resultados
print(romi[['source_id', 'revenue', 'buyers', 'LTV', 'CAC', 'ROMI']])

# Crear el gr√°fico de barras
sns.set(style="whitegrid")

# Crear el gr√°fico de barras
plt.figure(figsize=(8, 4))
romi_plot = sns.barplot(
    data=romi.sort_values(by='ROMI', ascending=False),
    x='source_id',
    y='ROMI',
    palette='viridis'
)
plt.title('ROMI por Fuente de Adquisici√≥n', fontsize=14)
plt.xlabel('Fuente de Adquisici√≥n')
plt.ylabel('ROMI (LTV / CAC)')
plt.tight_layout()
plt.show()
```
